this library is a wrapper to run lucid with keras models. borrows a lot of func from lucid (https://github.com/tensorflow/lucid) and many codes as it is or partially rewritten for adaptation. If you with to use this library somewhere else, you first must cite and/or acknowledge Olah et al., 2017 Distil (https://distill.pub/2017/feature-visualization/) and its github repository and lucid4keras (https://github.com/totti0223/lucid4keras) as well.

```python
#Imports

import keras.backend as K
from keras.applications.inception_v3 import InceptionV3
from keras.applications.vgg16 import VGG16

import lucid.optvis.param as param
import lucid.optvis.transform as transform

from lucid4keras import prepare_model, keras_render_vis
from lucid4keras import objectives as keras_objectives

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

## Prepare the keras model

the original lucid package handles a graph that can search for a pre-relu layer however for keras, many users define the relu with the conv layers and have to linearize in prior to feeding to the render function.

the prepare_model function creates a intermediate output model with the defined layer_name and linearizes the final output (if already a linear, will do nothing). in case of mixed layers such included in inception modules, will search for the inbound layers and linearize that. if another mixed layer were connected, recursively find and linearize the inbound nodes.

```python
K.clear_session()
model = InceptionV3()
model = prepare_model(model,layer_name="mixed4")
```

## Visualize a Neuron

```python
# Visualizing a neuron is easy using keras too.
# original lucid accepts layername + neuron index for objectives but
# this wrapper accepts neuron index only since the layer is specified already above.

images = keras_render_vis(model,127)

#returns numpy array of (1,128,128,3) by default
```

```python
#can also draw or save with matplot function

plt.figure(figsize=(2,2))
plt.imshow(images[0])
```

```python
#changing iteration number and obtaining intermediate iterations with thresholds.
#iterates until the maximum threshold value
images = keras_render_vis(model,"127",thresholds=[128,256,512])

#now images returned is a numpy array of (3,128,128,3) 
```

```python
#can also be handled by matplot library
print(images.shape)

for image in images:
    plt.figure(figsize=(2,2))
    plt.imshow(image)
    plt.show()
```

```python
#outputs durring the calculation can be supressed by verbose

images = keras_render_vis(model,127,thresholds=[128,256,512],verbose=False)
print(image.shape)
```

## Deeper

**Experimenting with objectives**

```python
# with more explicit objective:

obj = keras_objectives.channel(127)
image = keras_render_vis(model, obj)
```

```python
channel = lambda n: keras_objectives.channel(n)
obj = channel(127)
image = keras_render_vis(model, obj)

obj = channel(128)
image = keras_render_vis(model, obj)

obj = channel(127) + channel(128)
image = keras_render_vis(model, obj)
```

**Transormation Robustness**

#by default it uses the lucid default transformation 

from lucid.optvis.transform.py

`standard_transforms = [
    pad(12, mode="constant", constant_value=.5),   
    jitter(8),
    random_scale([1 + (i - 5) / 50. for i in range(11)]),
    random_rotate(list(range(-10, 11)) + 5 * [0]),
    jitter(4),
]`

lets control it manualy

```python
#transformation completely off by passing an empty list.
#transforms = None (which is default) applies standard transformation
#so becarefull what you are doing
image = keras_render_vis(model, 127, transforms = [])
```

```python
# jitter 2

transforms = [
    transform.jitter(2)
]

image = keras_render_vis(model, 127, transforms = transforms)
```

**Experimenting with parameterization**

```python
# Using alternate parameterizations is one of the primary ingredients for
# effective visualization

param_f = lambda: param.image(128, fft=False, decorrelate=False)
images = keras_render_vis(model, 127, param_f)
```

```python
param_f = lambda: param.image(128, fft=True, decorrelate=True)
images = keras_render_vis(model, 127, param_f)
```

```python
# change input image size

param_f = lambda: param.image(299)
images = keras_render_vis(model, 80, param_f,thresholds=[128,256,512])
```

## Negative Neurons ( and Batch processing)

```python
#See the negative neurons (image that neurons dislike (or rather shows no interest at all))

channel = lambda n: keras_objectives.channel(n)
obj = channel(127)
image = keras_render_vis(model, obj)

obj = -channel(127)
image = keras_render_vis(model, obj)
```

```python
#note that without batch dimension or on same batch number, it will be a adding or subtraction of neurons.
#however the below differs in batch dimention so will be a stacked numpy of 
#1. positive 127 neuron and 2. negative 127 neuron
obj = keras_objectives.channel(127,batch=0) - keras_objectives.channel(127,batch=1)
image = keras_render_vis(model, obj)
```

## Random Directions

```python
#input have to be equal to the number of neurons in the layer of interest. counting by the below command
no_neurons = model.get_layer("mixed4").output_shape[3]

obj = keras_objectives.direction(np.random.randn(no_neurons))

param_f = lambda: param.image(224)

images = keras_render_vis(model, obj, param_f=param_f, thresholds=[256,512,1024,2048])
```
