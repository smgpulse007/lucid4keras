this library is a wrapper to run lucid with keras models. borrows a lot of func from lucid (https://github.com/tensorflow/lucid) and many codes as it is or partially rewritten for adaptation. If you with to use this library somewhere else, you first must cite and/or acknowledge Olah et al., 2017 Distil (https://distill.pub/2017/feature-visualization/) and its github repository and lucid4keras (https://github.com/totti0223/lucid4keras) as well.

```python
#Imports

import keras.backend as K
from keras.applications.inception_v3 import InceptionV3
from keras.applications.vgg16 import VGG16

import lucid.optvis.param as param
import lucid.optvis.transform as transform

from lucid4keras import prepare_model, keras_render_vis
from lucid4keras import objectives as keras_objectives

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
K.clear_session()
model = InceptionV3()
model = prepare_model(model,layer_name="mixed4")
```

## Interpolating between neurons

```python
#customized function for defining number of images for interpolation
#i do not use the param.lowres_tensor used in lucid as i don't fully understand what they are doing. still works without it.
#https://colab.research.google.com/github/tensorflow/lucid/blob/master/notebooks/feature-visualization/neuron_interaction.ipynb#scrollTo=GYujByXLdap7

#neuron1, neuron2, and number of images to interpolate
obj = keras_objectives.channel_interpolate(127,128,7)

param_f = lambda: param.image(224)

images = keras_render_vis(model, obj, param_f,thresholds=[256,512])
```

```python
#lets compare with the simple sum of two neurons
#should look like the intermediate image above
channel = lambda n,batch: keras_objectives.channel(n,batch)

obj = channel(127, batch=0) + (channel(127, batch = 1) + channel(128, batch = 1)) + channel(128, batch = 2)
print(obj.batch_n)
param_f = lambda: param.image(224)

images = keras_render_vis(model, obj, param_f,thresholds=[256,512])
```
